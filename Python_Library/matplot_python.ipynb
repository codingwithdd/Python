```python
slices=[3,5,8,9,1]
```

```python
activities=["Sleeping","Eating","Enjoying","Working","riding"]
```

```python
cols=["c","m","r","b","y"]
```

```python
plt.pie(slices,
       labels=activities,
       colors=cols,
       startangle=180, # starting slice vertically-90, horizontally-180
       shadow=True,
       explode=(0,0.1,0,0,0), # increase points to looks like ate
       autopct='%1.1f%%') # gives percentage over slice
       
plt.title("Pie chart") # title to piechart


plt.show()
```

***Working with multiple plots***

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1= np.arange(0,5,0.1)
t2= np.arange(0,5,0.02)

plt.subplot(221) 
# first 2-two graphs avilable 
#second 2- two herizontal graph 
#third 1- among two graph (this is first)
plt.plot(t1,f(t1)),"bo",t2,f(t2)

plt.subplot(222)
# first 2-two graphs avilable 
#second 2- two herizontal graph 
#third 2- among two graph (this is second)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
